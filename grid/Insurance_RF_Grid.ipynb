```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'criterion':['squared_error', 'absolute_error', 'friedman_mse', 'poisson'],
            'max_features':['sqrt', 'log2'],'n_estimators':[50,100]}
from sklearn.ensemble import RandomForestRegressor
grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
grid_results=grid.cv_results_
grid_results
table=pd.DataFrame.from_dict(grid_results)
table
```

```python
y_predict=grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2score=r2_score(y_test,y_predict)
```

```python
print(f"best parameter{grid.best_params_} \nr2score value:{r2score}")
```
